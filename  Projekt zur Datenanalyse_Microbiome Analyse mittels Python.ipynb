```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from sklearn.impute import SimpleImputer
```

* Data characteristics

```python
df=pd.read_csv('datasets\chronic_bowl.csv')
df
```

#### Split our data into input_variable and Output_variable
* we have 44 features 
* target_variable, is ibd_subtype . with 7 categories  ['CCD', 'UC', 'ICD_r', 'ICD_nr', 'HC', 'CC', 'LC']

```python
X=df.drop('ibd_subtype',axis=1)
Y=df['ibd_subtype']
```

#### Handling missing data
* Handling missing data in **bmi** with median
* Using **Ordinal Encoding** for target variable
    * assigns each unique Subtype of IBD to a different integer[1-6]
*Using **one hot encoding** to handle categorical value in inputs Variable
    * creates new columns indicating the presence (or absence) of each possible value in the original data
    
    

```python
X=X.replace('missing: not provided',np.NAN)
X['bmi']=X['bmi'].astype("float64")
m=round(X['bmi'].median(),2)
X['bmi'].fillna(m,inplace=True)
X['bmi'].unique()
```

```python
#Median=26
X['bmi'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder   
labelencoder_Y=LabelEncoder()
Y=labelencoder_Y.fit_transform(Y)
```

```python
s=(X.dtypes=='object')
object_cols = list(s[s].index)
```

```python
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(X[object_cols]))
OH_cols_train.index = X.index
num_X_train = X.drop(object_cols, axis=1)
OH_X_train = pd.concat([num_X_train, OH_cols_train], axis=1)
```

```python
#X['diagnosis'] = cancer_data['diagnosis'].map({'M':1,'B':0})
OH_X_train['dna_extracted']=OH_X_train['dna_extracted'].map({True:1})
OH_X_train['physical_specimen_remaining']=OH_X_train['physical_specimen_remaining'].map({False:0})

for col in  OH_X_train:
    print(col,OH_X_train[col].unique())
```
